```python
# Se dau mai multe imagini (salvate in folder-ul "data/images"). Se cere:

# - sa se vizualizeze una din imagini
# - daca imaginile nu aceeasi dimensiune, sa se redimensioneze toate la 128 x 128 pixeli si sa se vizualizeze imaginile intr-un cadru tabelar.
# - sa se transforme imaginile in format gray-levels si sa se vizualizeze
# - sa se blureze o imagine si sa se afiseze in format "before-after"
# - sa se identifice muchiile ontr-o imagine si sa se afiseze in format "before-after"
```

```python
# sa se vizualizeze una din imagini
from PIL import Image

image = Image.open("data/images/Norvig.jpg")
image.show()
```

```python
# daca imaginile nu aceeasi dimensiune, sa se redimensioneze toate la 128 x 128 pixeli si sa se vizualizeze imaginile intr-un cadru tabelar.
import os
import matplotlib.pyplot as plt

folder_path = "data/images/"
resized_images = []

for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)
    
    image = Image.open(file_path)
    resized_image = image.resize((128, 128))
    resized_images.append(resized_image)

# Calculm numrul de rnduri i coloane necesare pentru afiarea imaginilor ntr-un cadru tabelar
num_images = len(resized_images)
num_rows = (num_images // 3) + (1 if num_images % 3 != 0 else 0)
num_cols = min(3, num_images)

# Configurm subgraficele folosind matplotlib
fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 8))

# Parcurgem fiecare imagine redimensionat i o afim n subgrafic
for i, ax in enumerate(axes.flat):
    if i < num_images:
        ax.imshow(resized_images[i])
        ax.axis('off')  # Dezactivm axele
    else:
        ax.axis('off')  # Dezactivm axele pentru subgraficele neutilizate

# Ajustm aspectul subgraficelor i afim imaginea
plt.tight_layout()
plt.show()
```

```python
# sa se transforme imaginile in format gray-levels si sa se vizualizeze
for img in resized_images:
    # Convertii imaginea n nivele de gri
    image_gray = img.convert("L")
    image_gray.show()
```

```python
# sa se blureze o imagine si sa se afiseze in format "before-after"
from PIL import ImageFilter

image_blurred = image.filter(ImageFilter.BLUR)

# Afiai imaginea original i imaginea blurrat n format "before-after"
combined_image = Image.new("RGB", (image.width * 2, image.height))
combined_image.paste(image, (0, 0))
combined_image.paste(image_blurred, (image.width, 0))

combined_image.show()
```

```python
# sa se identifice muchiile intr-o imagine si sa se afiseze in format "before-after"

# Detecting Edges on the Image using the argument ImageFilter.FIND_EDGES
image = Image.open("data/images/Norvig.jpg")
image_edges = image.filter(ImageFilter.FIND_EDGES)

combined_image = Image.new("RGB", (image.width * 2, image.height))
combined_image.paste(image, (0, 0))
combined_image.paste(image_edges, (image.width, 0))
 
combined_image.show()
```
